```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
header  = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
data    = pd.read_csv("data/iris.data", names=header)

classes = data["class"].unique()
sizes   = [sum(data["class"] == c) for c in classes]
mask_cls = [data["class"] == c for c in classes]
```

## 1) Sepal and Petal

```python
colors = ["red", "blue", "green"]
labels = ["Sepal", "Petal"]
ax = [plt.subplots()[1], plt.subplots()[1]]

for i, cls in enumerate(mask_cls):
    ax[0].scatter(data[cls]["sepal_length"], data[cls]["sepal_width"], lw=2, label=classes[i], color=colors[i])
    ax[1].scatter(data[cls]["petal_length"], data[cls]["petal_width"], lw=2, label=classes[i], color=colors[i])

for i in range(2):
    ax[i].legend(loc="upper right")
    ax[i].set_xlabel("Length")
    ax[i].set_ylabel("Width")
    ax[i].set_title(labels[i])
```

## 2) Lenght-Lenght, Width-Width

```python
colors = ["red", "blue", "green"]
labels = ["Length", "Width"]
ax = [plt.subplots()[1], plt.subplots()[1]]

for i, cls in enumerate(mask_cls):
    ax[0].scatter(data[cls]["sepal_length"], data[cls]["petal_length"], lw=2, label=classes[i], color=colors[i])
    ax[1].scatter(data[cls]["sepal_width"], data[cls]["petal_width"], lw=2, label=classes[i], color=colors[i])

for i in range(2):
    ax[i].legend(loc="upper right")
    ax[i].set_xlabel("Sepal")
    ax[i].set_ylabel("Petal")
    ax[i].set_title(labels[i])
```

## 3) Length-Width (not so intuitive)

```python
colors = ["red", "blue", "green"]
labels = ["length", "width"]
ax = [plt.subplots()[1], plt.subplots()[1]]

for i, cls in enumerate(mask_cls):
    ax[0].scatter(data[cls]["sepal_length"], data[cls]["petal_width"], lw=2, label=classes[i], color=colors[i])
    ax[1].scatter(data[cls]["sepal_width"], data[cls]["petal_length"], lw=2, label=classes[i], color=colors[i])

for i in range(2):
    ax[i].legend(loc="upper right")
    ax[i].set_xlabel("Sepal "+labels[i])
    ax[i].set_ylabel("Petal "+labels[i])
    ax[i].set_title("")
```
